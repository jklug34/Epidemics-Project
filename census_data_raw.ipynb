```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census
from us import states
import gmaps

# Census & gmaps API Keys
from config import (api_key, gkey)
c = Census(api_key, year= 2013)

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Run Census Search to retrieve data on all states
census_data = c.acs5.get(("NAME", "B19013_001E", "B01003_001E", "B01002_001E", "B19301_001E", "B17001_002E", "B23025_005E", "B15003_002E", "B15003_017E", "B15003_018E", "B15003_021E", "B15003_022E", "B15003_023E", "B15003_024E", "B15003_025E", "B02001_002E", "B02001_003E", "B02001_004E", "B02001_005E", "B02001_006E", "B02001_008E", "B03001_003E"), {'for': 'state:*'})
#"B15003_002E"

# Make dataframe
census_pd = pd.DataFrame(census_data)
census_pd.head()
```

```python
# Reorder columns 
census_pd = census_pd.rename(columns={"B01003_001E": "Total Population",
                                      "B01002_001E": "Median Age",
                                      "B19013_001E": "Median Household Income Per Year (Dollars)",
                                      "B19301_001E": "Income Per Capita Per Year (Dollars)",
                                      "B17001_002E": "Number of People Living in Poverty",
                                      "B23025_005E": "Unemployment Count (16+)",
                                      "B15003_002E": "Number of People Who Completed No Schooling (25+)",
                                      "B15003_017E": "Number of People with High School Degree (25+)",
                                      "B15003_018E": "Number of People with GED (25+)",
                                      "B15003_021E": "Number of People with Associates Degree (25+)",
                                      "B15003_022E": "Number of People with Bachelors Degree (25+)",
                                      "B15003_023E": "Number of People with Masters Degree (25+)",
                                      "B15003_024E": "Number of People with Professional Degree (25+)",
                                      "B15003_025E": "Number of People with Doctoral Degree (25+)",
                                      "B02001_002E": "Number of People of European American Origin",
                                      "B02001_003E": "Number of People of African American Origin",
                                      "B02001_004E": "Number of People of Native American Origin",
                                      "B02001_005E": "Number of People of Asian Origin",
                                      "B02001_006E": "Number of People of Pacific Islander Origin",
                                      "B03001_003E": "Number of People of Hispanic Origin",
                                      "B02001_008E": "Number of People of Multiethnic Origin",
                                      "NAME": "Name", "state": "State"  
})

census_pd.head()
```

```python
# Add in Percentage of People Living in Poverty (Number of People Living in Poverty / Total Population)
census_pd["Percentage of People Living in Poverty"] = (census_pd["Number of People Living in Poverty"].astype(int) / census_pd["Total Population"].astype(int)) * 100
```
